```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
print(X_train)
print(Y_train)
print(X_test)
print(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scx=StandardScaler()
X_train=scx.fit_transform(X_train)
X_test=scx.transform(X_test)
X_train
X_test
```

```python
scy=StandardScaler()
Y_train=scy.fit_transform(Y_train)
Y_test=scy.transform(Y_test)
Y_train
Y_test
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

param_grid={'criterion':['squared_error','absolute_error','friedman_mse'],
           'max_features':[None,'sqrt','log2'],
           'splitter':['best','random']}
grid=GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
print(re)
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
grid.best_params_
```

```python
Y_predict=grid.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import r2_score
r2=r2_score(Y_test,Y_predict)
r2
```

```python
preinput=scx.transform([[40,28.5,2,1,0]])
preoutput=grid.predict(preinput)
result=scy.inverse_transform([preoutput])
result
```

```python
import pickle
pickle.dump(grid,open('finalized_DesisionTreeWithoutSavingSS_model.sav','wb'))
pickle.dump(scx,open('standardScalerX.sav','wb'))
pickle.dump(scy,open('standardScalerY.sav','wb'))
```

```python
loaded_model=pickle.load(open('finalized_DesisionTreeWithoutSavingSS_model.sav','rb'))
loaded_scx=pickle.load(open('standardScalerX.sav','rb'))
loaded_scy=pickle.load(open('standardScalerY.sav','rb'))

preinput=loaded_scx.transform([[40,28.5,2,1,0]])
preoutput=loaded_model.predict(preinput)
result=loaded_scy.inverse_transform([preoutput])
result
```
